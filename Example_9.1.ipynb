```python
def koch_snowflake(order, scale=10):

    """
    Return two lists x, y of point coordinates of the Koch snowflake.
    Parameters
    ----------
    order : int
        The recursion depth.
    scale : float
        The extent of the snowflake (edge length of the base triangle).
    """

    def _koch_snowflake_complex(order):
        if order == 0:

            # initial triangle
            angles = np.array([0, 120, 240]) + 90
            return scale / np.sqrt(3) * np.exp(np.deg2rad(angles) * 1j)

        else:

            ZR = 0.5 - 0.5j * np.sqrt(3) / 3
            p1 = _koch_snowflake_complex(order - 1)  # start points
            p2 = np.roll(p1, shift=-1)  # end points
            dp = p2 - p1  # connection vectors
            new_points = np.empty(len(p1) * 4, dtype=np.complex128)
            new_points[::4] = p1
            new_points[1::4] = p1 + dp / 3
            new_points[2::4] = p1 + dp * ZR
            new_points[3::4] = p1 + dp / 3 * 2

            return new_points

    points = _koch_snowflake_complex(order)
    x, y = points.real, points.imag
    return x, y

x, y = koch_snowflake(order=5)
plt.fill(x, y)
plt.axis('equal')
plt.show()
```

```python
plt.fill(x-2, y-2, color='red', alpha=0.5)
plt.fill(x+2, y+2, color='green', alpha=0.5)
plt.fill(x, y, color='blue', alpha=0.5)
plt.axis('equal')
plt.show()
```

### Podejście obiektowe

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])  # left, bottom, width, height

x = np.arange(11)
y = x ** 2
axes.plot(x, y)
axes.set_xlabel('oś X')
axes.set_ylabel('oś Y')
axes.set_title('Tytuł wykresu')
plt.show()
```

```python
rand_arr = np.random.rand(10, 2)  # Tworzymy tablicę z losowymi punktami

fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])  # Pierwsze osie (cała figura)
axes2 = fig.add_axes([0.2, 0.2, 0.8, 0.8])  # Drugie osie (mniejszy wykres wewnątrz)

x = np.arange(11)
y = x ** 2
axes1.plot(x, y)  # Dodaj wykres do axes1
axes2.scatter(rand_arr[:, 0], rand_arr[:, 1])  # Dodaj wykres punktowy do axes2
plt.show()
```

### Formatowanie wyglądu wykresów

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='x^2')
axes.plot(x,x**3,label='x^3')

axes.legend(loc=0)
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='x^2')
axes.plot(x,x**3,label='x^3')

axes.legend(loc=(1.05,0.0))
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='x^2',color='red',linewidth=3)
axes.plot(x,x**3,label='x^3',linewidth=3,alpha=0.3)
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='x^2',color='red',linewidth=3,linestyle='--')
axes.plot(x,x**3,label='x^3',linewidth=3,linestyle='-.')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='x^2',
          color='red',
          linewidth=3,
          marker='o',
          markersize=20,
          markerfacecolor='orange',
          markeredgewidth=2)
axes.plot(x,x**3,label='x^3',linewidth=3,
          marker='*',
          markersize=10)
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.yaxis.set_major_locator(plt.NullLocator())
axes.xaxis.set_major_formatter(plt.NullFormatter())

axes.plot(x,x**2,label='x^2')
axes.plot(x,x**3,label='x^3')
plt.show()
```

### Wbudowana wizualizacja danych w Pandas

```python
import pandas as pd
import datetime as dt
df = pd.DataFrame({'A':np.random.randn(200),
                 'B':[dt.datetime(2019,1,1)+dt.timedelta(days=x) for x in range(200)],
                  'C':np.arange(1,201)})

df['A'].hist(bins=30)
plt.show()
```

```python
df['A'].plot(kind='hist',bins=30)
plt.show()
```

```python
df['C'].loc[:10].plot(kind='bar')
plt.show()
```

```python
df.plot.line(x='B',y='A',figsize=(12,5), c='red',lw=2)
plt.show()
```
